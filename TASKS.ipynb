```python
import pandas as pd
from statistics import median, mode


csv_file_path = '/content/sample_data/california_housing_test.csv'
data = pd.read_csv('/content/sample_data/california_housing_test.csv')


results = {}

for column in data.columns:
    column_data = data[column]
    column_mean = column_data.mean()
    column_median = median(column_data)
    column_mode = mode(column_data)

    results[column] = {
        'mean': column_mean,
        'median': column_median,
        'mode': column_mode
    }

# Print the results
for column, stats in results.items():
    print(f"Column: {column}")
    print(f"Mean: {stats['mean']}")
    print(f"Median: {stats['median']}")
    print(f"Mode: {stats['mode']}")
    print("-" * 20)
```

```python
import random
import pandas as pd
import matplotlib.pyplot as plt

# Function to play the dice rolling game
def play_dice_game(player_choice):
    player_roll = random.randint(1, 6)
    computer_roll = random.randint(1, 6)

    if player_roll > computer_roll:
        return 'player'
    elif player_roll < computer_roll:
        return 'computer'
    else:
        return 'tie'

# Main game loop
player_wins = 0
computer_wins = 0
total_games = 0

while True:
    user_input = input("Press 'r' to roll the dice or 'q' to quit: ").lower()

    if user_input == 'q':
        break
    elif user_input != 'r':
        print("Invalid input. Please press 'r' to roll or 'q' to quit.")
        continue

    result = play_dice_game(user_input)

    if result == 'player':
        player_wins += 1
        print("You win!")
    elif result == 'computer':
        computer_wins += 1
        print("Computer wins!")
    else:
        print("It's a tie!")

    total_games += 1
    print(f"Player wins: {player_wins} | Computer wins: {computer_wins}\n")

# Save results to Excel
results_dict = {'Player Wins': [player_wins], 'Computer Wins': [computer_wins]}
results_df = pd.DataFrame(results_dict)

results_df.to_excel('dice_game_results.xlsx', index=False)
print("Results saved to 'dice_game_results.xlsx'")

# Calculate success rates
player_success_rate = player_wins / total_games
computer_success_rate = computer_wins / total_games

# Plot the success rates
labels = ['Player', 'Computer']
success_rates = [player_success_rate, computer_success_rate]
colors = ['green', 'blue']

plt.bar(labels, success_rates, color=colors)
plt.xlabel('Participants')
plt.ylabel('Success Rate')
plt.title('Success Rate of Player vs. Computer')
plt.ylim(0, 1)  # Set y-axis range between 0 and 1

for i, rate in enumerate(success_rates):
    plt.text(i, rate + 0.05, f'{rate:.2f}', ha='center', color='black', fontsize=12)

plt.show()
```
