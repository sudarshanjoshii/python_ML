```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=np.array([1,5,13,25,33,46]).reshape(-1,1)
y=np.array([3,6,15,12,21,18])
plt.scatter(x,y)
plt.xlabel("IV")
plt.ylabel("DV")
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
SLR=LinearRegression()
SLR.fit(x,y)
```

```python
pred=SLR.predict(x)
```

```python
from sklearn.metrics import mean_squared_error
mse=mean_squared_error(y,pred)
plt.plot(x,pred,color='red',marker='*')
print("MSE:",mse)
print("intercept",SLR.intercept_)
print("slope:",SLR.coef_)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
import random

random.seed(42)

num_samples = 8

x = np.array([random.uniform(1, 50) for _ in range(num_samples)]).reshape((-1, 1))
y = np.array([2 * val + random.uniform(-5, 5) for val in x])


plt.scatter(x, y)
plt.xlabel("IV")
plt.ylabel("DV")
plt.title("Scatter Plot of Data")
plt.show()

SLR = LinearRegression()
SLR.fit(x, y)

pred = SLR.predict(x)

mse = mean_squared_error(y, pred)

plt.scatter(x, y, label="Actual Data")
plt.plot(x, pred, color='orange', label="Regression Line")
plt.xlabel("IV")
plt.ylabel("DV")
plt.title("Linear Regression Fit")
plt.legend()
plt.show()

print("MSE:", mse)
print("Intercept:", SLR.intercept_)
print("Slope:", SLR.coef_)
```
